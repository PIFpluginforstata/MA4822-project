```python
#!/usr/bin/env python3
"""
Multi-Axis Force/Torque Sensor ML Calibration System
Version: 2.4 - Jupyter Notebook Optimized
"""
import os

# !!! Thread settings MUST be at the very beginning !!!
os.environ['OMP_NUM_THREADS'] = '22'
os.environ['MKL_NUM_THREADS'] = '22'
os.environ['OPENBLAS_NUM_THREADS'] = '22'

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import pandas as pd
import time
import random
from dataclasses import dataclass, field
from typing import Tuple, List, Dict
from sklearn.model_selection import KFold

# Global Random Seed
np.random.seed(0)
torch.manual_seed(0)
random.seed(0)

# Set PyTorch threads (Only setting this works in Jupyter)
torch.set_num_threads(22)


@dataclass
class SensorConfig:
    sample_rate: float = 1000.0
    temp_range: Tuple[float, float] = (0, 50)
    force_range: np.ndarray = field(default_factory=lambda: np.array([130.0, 130.0, 400.0]))
    torque_range: np.ndarray = field(default_factory=lambda: np.array([10.0, 10.0, 10.0]))
    noise_sigma: float = 0.05
    creep_C_range: Tuple[float, float] = (0.015, 0.035)
    creep_tau_range: Tuple[float, float] = (1.0, 10.0)


class SensorDataGenerator:
    """Data Generator (Includes Creep Model)"""
    def __init__(self, cfg: SensorConfig):
        self.cfg = cfg
        self.C_true = np.eye(6) + (np.random.rand(6, 6) - 0.5) * 0.05
        self.Z_true = (np.random.rand(6, 4) - 0.5) * np.array([0.1, 0.05, 1e-3, 1e-5])[None, :]
        self.S_true = np.column_stack([np.ones(6) + (np.random.rand(6) - 0.5) * 0.02,
                                       (np.random.rand(6) - 0.5) * 2e-3])
        self.true_creep_C = np.random.uniform(*cfg.creep_C_range)
        self.true_creep_tau = np.random.uniform(*cfg.creep_tau_range)
        print(f"[Generator] True Creep C={self.true_creep_C:.4f}, Tau={self.true_creep_tau:.2f}s")

    def _loads(self, t, mode):
        n = len(t)
        f = np.zeros((n, 6))
        if mode == "static":
            f += np.random.normal(0, 0.02, (n, 6))
        elif mode == "dynamic":
            f[:, 0] = 50 * np.sin(0.1 * 2 * np.pi * t)
            f[:, 1] = 30 * np.sin(0.2 * 2 * np.pi * t + np.pi / 4)
            f[:, 2] = 100 * np.sin(0.05 * 2 * np.pi * t)
            f[:, 3] = 2 * np.sin(0.15 * 2 * np.pi * t)
            f[:, 4] = 1.5 * np.sin(0.1 * 2 * np.pi * t + np.pi / 3)
            f[:, 5] = 3 * np.sin(0.08 * 2 * np.pi * t)
        else:  # "complex"
            tm = t / (t[-1] if t[-1] > 0 else 1.0)
            f[:, 0] = 50 + 20 * np.sin(0.1 * 2 * np.pi * t) + 10 * np.random.randn(n) * (0.5 + tm)
            f[:, 1] = 30 * (1 + 0.5 * np.sin(0.05 * 2 * np.pi * t)) * (1 - 0.3 * tm)
            f[:, 2] = 100 + 50 * np.sign(np.sin(0.02 * 2 * np.pi * t))
            f[:, 3] = 2 * np.exp(-t / (t[-1] / 2 if t[-1] > 0 else 1)) * np.sin(0.2 * 2 * np.pi * t * (1 + 0.5 * tm))
            f[:, 4] = 1.5 + 5 * tm ** 2
            f[:, 5] = 3 * (1 + 0.3 * np.sin(0.15 * 2 * np.pi * t)) + np.cumsum(np.random.randn(n) * 0.01)
            f += np.cumsum(np.random.normal(0, 0.005, f.shape), axis=0)
        return f

    def _temp(self, t):
        base = 25 + 15 * np.sin(2 * np.pi * t / max(t[-1], 1.0))
        ac = 2 * np.sin(2 * np.pi * t / 600)
        rw = np.cumsum(np.random.normal(0, 0.01, len(t)))
        if len(rw) > 1:
            rw = rw - np.linspace(rw[0], rw[-1], len(rw))
        x = base + ac + rw
        return np.clip(x, 5, 45)

    def _forward_with_creep(self, F: np.ndarray, T: np.ndarray, t: np.ndarray) -> np.ndarray:
        """Forward Model with Creep"""
        n_samples = F.shape[0]
        V = np.zeros_like(F)
        
        if n_samples == 0:
            return V
            
        dT = (T - 25.0)[:, None]
        Z = self.Z_true[:, 0] + self.Z_true[:, 1] * dT + self.Z_true[:, 2] * dT ** 2 + self.Z_true[:, 3] * dT ** 3
        S = self.S_true[:, 0] + self.S_true[:, 1] * dT
        V_ideal_no_creep = (F @ self.C_true.T) * S + Z
        
        dt = t[1] - t[0] if n_samples > 1 else 0.0
        t_creep_state = np.zeros(6)
        last_F = F[0, :]
        
        force_range_full = np.concatenate((self.cfg.force_range, self.cfg.torque_range))
        change_threshold = force_range_full * 0.005
        mag_threshold = force_range_full * 0.01

        for i in range(n_samples):
            creep_factor_vec = 1.0 + self.true_creep_C * np.log(1 + t_creep_state / (self.true_creep_tau + 1e-9))
            V[i, :] = V_ideal_no_creep[i, :] * creep_factor_vec

            if i < n_samples - 1:
                current_F = F[i, :]
                F_change = np.abs(current_F - last_F)
                F_mag = np.abs(current_F)
                
                for j in range(6):
                    if F_change[j] < change_threshold[j] and F_mag[j] > mag_threshold[j]:
                        t_creep_state[j] += dt
                    else:
                        t_creep_state[j] = 0.0
                
                last_F = current_F
                
        V += np.random.normal(0, self.cfg.noise_sigma, V.shape)
        return V

    def make(self, duration, mode):
        n = max(1, int(duration * self.cfg.sample_rate))
        t = np.linspace(0, duration, n)
        F = self._loads(t, mode)
        T = self._temp(t)
        V = self._forward_with_creep(F, T, t)
        
        df = pd.DataFrame({"time": t, "temp": T})
        for i in range(6):
            df[f"V{i + 1}"] = V[:, i]
        names = ["Fx_true", "Fy_true", "Fz_true", "Mx_true", "My_true", "Mz_true"]
        for i, nm in enumerate(names):
            df[nm] = F[:, i]
        return df


def calculate_t_creep(F_data: np.ndarray, t_data: np.ndarray, cfg: SensorConfig) -> np.ndarray:
    """Pre-calculates t_creep vector"""
    n_samples = F_data.shape[0]
    t_creep_vec = np.zeros_like(F_data)
    
    if n_samples == 0:
        return t_creep_vec
        
    dt = t_data[1] - t_data[0] if n_samples > 1 else 0.0
    t_creep_state = np.zeros(6)
    last_F = F_data[0, :]
    
    force_range_full = np.concatenate((cfg.force_range, cfg.torque_range))
    change_threshold = force_range_full * 0.005
    mag_threshold = force_range_full * 0.01

    for i in range(n_samples):
        t_creep_vec[i, :] = t_creep_state
        
        current_F = F_data[i, :]
        F_change = np.abs(current_F - last_F)
        F_mag = np.abs(current_F)

        for j in range(6):
            if F_change[j] < change_threshold[j] and F_mag[j] > mag_threshold[j]:
                t_creep_state[j] += dt
            else:
                t_creep_state[j] = 0.0
        
        last_F = current_F
        
    return t_creep_vec


class ForcesSensorCalibration(nn.Module):
    """74-Parameter PyTorch Model"""
    def __init__(self, C0=None, Z0=None, S0=None, creep_C0=0.025, creep_tau0=5.0):
        super().__init__()
        if C0 is None: C0 = torch.eye(6)
        if Z0 is None: Z0 = torch.zeros(6, 4)
        if S0 is None:
            S0 = torch.ones(6, 2)
            S0[:, 1] = 0.0
            
        self.C = nn.Parameter(C0.float().clone())
        self.Z = nn.Parameter(Z0.float().clone())
        self.S = nn.Parameter(S0.float().clone())
        
        self.creep_C = nn.Parameter(torch.tensor(float(max(1e-6, creep_C0))))
        self.creep_tau = nn.Parameter(torch.tensor(float(max(1e-6, creep_tau0))))

    def forward(self, F, T, t_creep):
        dT = (T - 25.0).unsqueeze(1)
        
        Z = self.Z[:, 0] + self.Z[:, 1] * dT + self.Z[:, 2] * dT ** 2 + self.Z[:, 3] * dT ** 3
        S = self.S[:, 0] + self.S[:, 1] * dT
        
        V_ideal = (F @ self.C.t()) * S + Z
        
        creep_factor = 1.0 + self.creep_C * torch.log(1 + t_creep / (self.creep_tau + 1e-9))
        creep_factor = torch.clamp(creep_factor, 0.5, 2.0)
        
        return V_ideal * creep_factor


def stage1_separate(F, T, V):
    """Stage 1: LSTSQ Separate Calibration"""
    T_ref = 25.0
    
    mask = (F.abs().sum(1) < 0.01)
    if mask.sum() < 10:
        mask = (F.abs().sum(1) < 0.1)
        
    Tz = (T[mask] - T_ref).unsqueeze(1)
    M = torch.cat([torch.ones_like(Tz), Tz, Tz ** 2, Tz ** 3], 1)
    
    if M.shape[0] < M.shape[1]:
        print("Stage-1 Warning: Insufficient zero-load points for Z calibration, using default Z=0.")
        Z = torch.zeros(6, 4, device=V.device)
    else:
        Z = torch.linalg.lstsq(M, V[mask]).solution.t()

    m = (T - T_ref).unsqueeze(1)
    Zroom = Z[:, 0] + Z[:, 1] * m + Z[:, 2] * m ** 2 + Z[:, 3] * m ** 3
    V_temp_comp = V - Zroom
    
    mask2 = ((T - T_ref).abs() < 2.0) & (~mask)
    if mask2.sum() < 10:
        mask2 = (~mask)
        
    if mask2.sum() < 10:
        print("Stage-1 Warning: Insufficient load points for C calibration, using default C=I.")
        C = torch.eye(6, device=V.device)
    else:
        C = torch.linalg.lstsq(F[mask2], V_temp_comp[mask2]).solution.t()

    S = torch.ones(6, 2, device=V.device)
    S[:, 1] = 0.0
    
    creep_C0_init = 0.025
    creep_tau0_init = 5.0
    
    return C, Z, S, creep_C0_init, creep_tau0_init


def train_stage2(C0, Z0, S0, creep_C0, creep_tau0, F, T, t_creep, V, device, epochs=600, lr=1e-3, batch=32768):
    """Stage 2: ML Training (Optimized for CPU Parallelism)"""
    model = ForcesSensorCalibration(C0, Z0, S0, creep_C0, creep_tau0).to(device)
    
    opt = torch.optim.Adam(model.parameters(), lr=lr)
    sch = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, patience=50, factor=0.5, verbose=False)
    
    n = F.shape[0]
    
    # Pre-generate random indices for all epochs
    print(f"[ML] Preparing {epochs} epochs of random indices...")
    indices_list = [torch.randperm(n, device=device) for _ in range(epochs)]
    
    for e in range(epochs):
        idx = indices_list[e]
        tot = 0.0
        nb = 0
        
        # Use a larger batch to utilize multiple cores
        for i in range(0, n, batch):
            b = idx[i:i + batch]
            
            # Forward pass
            Vp = model(F[b], T[b], t_creep[b])
            loss = nn.functional.mse_loss(Vp, V[b])
            
            # Backward pass
            opt.zero_grad(set_to_none=True)  # Faster gradient zeroing
            loss.backward()
            opt.step()
            
            tot += loss.item()
            nb += 1
            
        avg_loss = tot / nb
        sch.step(avg_loss)
        
        # Report progress every 10 epochs
        if e % 10 == 0 or e == epochs - 1:
            print(f"[ML] Epoch {e+1}/{epochs} RMSE={np.sqrt(avg_loss):.6f}")
            
        if avg_loss < 1e-8:
            print("[ML] Early stopping.")
            break
            
    return model


def rmse_table(yhat_list, y, labels):
    """Calculates RMSE table"""
    ch = [f"V{i + 1}" for i in range(6)]
    rows = []
    for k, yp in zip(labels, yhat_list):
        e = yp - y
        r = np.sqrt((e ** 2).mean(0))
        rows.append([k, *r.tolist(), r.mean()])
    df = pd.DataFrame(rows, columns=["Method", *ch, "Overall"])
    return df


def plot_overall(df):
    """Plots overall RMSE comparison"""
    df_agg = df.groupby("Method")["Overall"].mean().reset_index()
    df_agg = df_agg.sort_values("Overall")
    plt.figure(figsize=(6, 3))
    plt.bar(df_agg["Method"], df_agg["Overall"], alpha=0.8)
    plt.ylabel("Avg. RMSE (Volt)")
    plt.title("Overall Average RMSE (across K-Folds)")
    plt.tight_layout()
    plt.show()


def plot_per_channel(df):
    """Plots per-channel RMSE comparison"""
    ch = [c for c in df.columns if c.startswith("V")]
    df_agg = df.groupby("Method")[ch].mean().reset_index()
    methods = df_agg["Method"]
    
    x = np.arange(len(ch))
    n_methods = len(methods)
    w_total = 0.8
    w_bar = w_total / n_methods
    
    plt.figure(figsize=(9, 4))
    colors = plt.cm.get_cmap('viridis', n_methods)
    
    for i, meth in enumerate(methods):
        offset = (i - (n_methods - 1) / 2) * w_bar
        plt.bar(x + offset, df_agg.loc[df_agg["Method"] == meth, ch].values.ravel(),
                 width=w_bar, label=meth, alpha=0.9, edgecolor="white", color=colors(i))
                
    plt.xticks(x, ch)
    plt.ylabel("Avg. RMSE (Volt)")
    plt.title("Per-Channel Average RMSE (across K-Folds)")
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_timeseries(t, V_true, V_list, labels, ch=0, nshow=3000):
    """Plots time series comparison"""
    s = slice(0, min(nshow, len(t)))
    plt.figure(figsize=(12, 3))
    plt.plot(t[s], V_true[s, ch], lw=1.2, label="True")
    for y, l in zip(V_list, labels):
        plt.plot(t[s], y[s, ch], lw=1.0, ls="--", label=l)
    plt.xlabel("Time (s)")
    plt.ylabel(f"V{ch + 1} (V)")
    plt.title(f"Channel V{ch+1} Time Series (Example from Fold 5)")
    plt.legend()
    plt.tight_layout()
    plt.show()


def plot_error_timeseries(t, V_true, V_list, labels, ch=0, nshow=3000):
    """Plots error time series"""
    s = slice(0, min(nshow, len(t)))
    plt.figure(figsize=(12, 3))
    for y, l in zip(V_list, labels):
        e = y[s, ch] - V_true[s, ch]
        plt.plot(t[s], e, lw=1.0, label=f"{l} Error")
    plt.axhline(0, color="black", lw=0.8, ls=":")
    plt.xlabel("Time (s)")
    plt.ylabel("Error (V)")
    plt.title(f"Channel V{ch+1} Error Time Series (Example from Fold 5)")
    plt.legend()
    plt.tight_layout()
    plt.show()


def main():
    DEVICE = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
    
    # Parameters for optimized CPU parallel performance
    N_SAMPLES_TOTAL = 400_000
    EPOCHS = 600
    BATCH = 32768  # Increase batch size to fully utilize 22-core CPU
    N_SPLITS = 5

    # Print system information
    print("="*60)
    print(f"CPU Mode (PyTorch Threads: {torch.get_num_threads()})")
    print(f"OMP Threads: {os.environ.get('OMP_NUM_THREADS', 'default')}")
    print(f"Batch Size: {BATCH}")
    print("="*60)
        
    cfg = SensorConfig()
    gen = SensorDataGenerator(cfg)
    
    dur_total = N_SAMPLES_TOTAL / cfg.sample_rate
    
    # Data Generation
    print(f"\nGenerating {N_SAMPLES_TOTAL} calibration data points (static, complex)...")
    df_temp = gen.make(dur_total * 0.1, "static")
    df_load = gen.make(dur_total * 0.9, "complex")
    
    df_all = pd.concat([df_temp, df_load], ignore_index=True)
    print(f"Total samples generated: {len(df_all)}.")
    
    # Pre-calculate t_creep
    print("\nPre-calculating t_creep vector...")
    F_all_np = df_all[[f"{k}_true" for k in ["Fx", "Fy", "Fz", "Mx", "My", "Mz"]]].values
    T_all_np = df_all["temp"].values
    V_all_np = df_all[[f"V{i}" for i in range(1, 7)]].values
    t_all_np = df_all["time"].values
    
    t_creep_all_np = calculate_t_creep(F_all_np, t_all_np, cfg)
    print("t_creep calculation complete.")

    # Convert to Torch Tensors
    F_all = torch.tensor(F_all_np, dtype=torch.float32, device=DEVICE)
    T_all = torch.tensor(T_all_np, dtype=torch.float32, device=DEVICE)
    V_all = torch.tensor(V_all_np, dtype=torch.float32, device=DEVICE)
    t_creep_all = torch.tensor(t_creep_all_np, dtype=torch.float32, device=DEVICE)

    # K-Fold Cross-Validation
    kf = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
    
    all_rmse_results = []
    example_plot_data = {}
    
    t_start_cv = time.time()
    
    for fold, (train_idx, test_idx) in enumerate(kf.split(F_all)):
        print(f"\n{'='*60}")
        print(f"Fold {fold + 1}/{N_SPLITS}")
        print(f"{'='*60}")
        
        # Data Split
        F_train, F_test = F_all[train_idx], F_all[test_idx]
        T_train, T_test = T_all[train_idx], T_all[test_idx]
        V_train, V_test = V_all[train_idx], V_all[test_idx]
        t_creep_train, t_creep_test = t_creep_all[train_idx], t_creep_all[test_idx]

        # Stage-1 Initialization
        print("[Fold] Running Stage-1 LSTSQ...")
        C1, Z1, S1, creep_C1, creep_tau1 = stage1_separate(F_train, T_train, V_train)

        # Stage-2 ML Training
        print(f"[Fold] Running Stage-2 ML (Epochs={EPOCHS}, N_train={len(train_idx)})...")
        t0 = time.time()
        model = train_stage2(C1, Z1, S1, creep_C1, creep_tau1,
                             F_train, T_train, t_creep_train, V_train, 
                             DEVICE, epochs=EPOCHS, lr=1e-3, batch=BATCH)
        print(f"[Fold] Training Time: {time.time() - t0:.2f}s")
        
        # Evaluation
        print("[Fold] Evaluating on test set...")
        model.eval()
        with torch.no_grad():
            F_test_cpu = F_test.cpu()
            T_test_cpu = T_test.cpu()
            t_creep_test_cpu = t_creep_test.cpu()
            V_test_cpu = V_test.cpu().numpy()
            
            V_uncal = (F_test_cpu @ torch.eye(6)).numpy()
            V_s1 = ForcesSensorCalibration(C1.cpu(), Z1.cpu(), S1.cpu(), 
                                           creep_C1, creep_tau1)(F_test_cpu, T_test_cpu, t_creep_test_cpu).numpy()
            V_s2 = model.cpu()(F_test_cpu, T_test_cpu, t_creep_test_cpu).numpy()
            V_true = V_test_cpu

        labels = ["Uncalibrated", "Separate (Stage-1)", "ML Global (Stage-2)"]
        df_rmse_fold = rmse_table([V_uncal, V_s1, V_s2], V_true, labels)
        df_rmse_fold['fold'] = fold
        all_rmse_results.append(df_rmse_fold)

        print("\n" + df_rmse_fold.to_string(index=False))

        # Save data from the last fold for plotting
        if fold == N_SPLITS - 1:
            example_plot_data = {
                't': t_all_np[test_idx],
                'V_true': V_true,
                'V_list': [V_uncal, V_s1, V_s2],
                'labels': labels
            }
            print("\n--- Final Model Parameters (Fold 5/5) ---")
            print(f"C Condition Number: {np.linalg.cond(model.C.detach().cpu().numpy()):.2f}")
            print(f"Learned Creep C: {model.creep_C.item():.6f} (True: {gen.true_creep_C:.6f})")
            print(f"Learned Creep Tau: {model.creep_tau.item():.4f}s (True: {gen.true_creep_tau:.4f}s)")

    print(f"\n{'='*60}")
    print(f"CV Training Complete, Total Time: {time.time() - t_start_cv:.2f}s")
    print(f"{'='*60}")
    
    # Aggregate K-Fold Results
    df_rmse_all = pd.concat(all_rmse_results, ignore_index=True)
    
    df_agg = df_rmse_all.drop(columns='fold').groupby('Method').mean()
    df_std = df_rmse_all.drop(columns='fold').groupby('Method').std()

    pd.set_option("display.float_format", lambda x: f"{x:.6f}")
    
    print("\n--- Average RMSE (5 Folds) ---")
    print(df_agg.to_string())
    
    print("\n--- RMSE Standard Deviation (5 Folds) ---")
    print(df_std.to_string())

    # Plotting
    plot_overall(df_agg.reset_index())
    plot_per_channel(df_agg.reset_index())
    
    if example_plot_data:
        plot_timeseries(example_plot_data['t'], 
                        example_plot_data['V_true'], 
                        example_plot_data['V_list'], 
                        example_plot_data['labels'], ch=0, nshow=4000)
        
        plot_error_timeseries(example_plot_data['t'], 
                              example_plot_data['V_true'], 
                              example_plot_data['V_list'], 
                              example_plot_data['labels'], ch=0, nshow=4000)


if __name__ == "__main__":
    main()
```
